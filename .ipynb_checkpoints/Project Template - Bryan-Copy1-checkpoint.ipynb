```python
data = "./resources/Spotify_Youtube.csv"
spot_yt_df = pd.read_csv(data)
spot_yt_df.head()
```

```python
#15 most danceable and streamed on YT and Spot
#15 most overall popular songs on spot and YT
#data frame for danceability against popularity - Coble
dance_pop_df = spot_yt_df[["Artist", "Track", "Album", "Danceability", "Views", "Stream"]]
dance_pop_df
```

```python
#YT views vs spotify streams - Bryan
yt_vs_spot_streams = spot_yt_df[["Artist", "Track", "Album", "Views", "Stream"]]

yt_vs_spot_streams["Views"] = yt_vs_spot_streams["Views"].fillna(0)
yt_vs_spot_streams["Stream"] = yt_vs_spot_streams["Stream"].fillna(0)

# Add "Views" and "Stream" together and create a new column "Combined Data"
#yt_vs_spot_streams["Combined Data"] = yt_vs_spot_streams["Views"] + yt_vs_spot_streams["Stream"]

yt_vs_spot_streams['Combined Views and Streams'] = (yt_vs_spot_streams['Stream'] + yt_vs_spot_streams['Views'])

yt_vs_spot_streams['Rank Views'] = yt_vs_spot_streams['Views'].rank(ascending=False, method='min')
yt_vs_spot_streams['Rank Stream'] = yt_vs_spot_streams['Stream'].rank(ascending=False, method='min')
yt_vs_spot_streams['Overall Rank'] = yt_vs_spot_streams['Combined Views and Streams'].rank(ascending=False, method='min')

# Sort the DataFrame by "Combined Views and Streams" in descending order
sorted_df = yt_vs_spot_streams.sort_values(by="Combined Views and Streams", ascending=False).reset_index(drop=True)

# Convert the DataFrame to string type
sorted_df = sorted_df.astype(str)

sorted_df
```
